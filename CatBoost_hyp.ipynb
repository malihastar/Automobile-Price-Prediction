```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.model_selection import  GridSearchCV
from sklearn import metrics
from catboost import CatBoostRegressor


df=pd.read_csv("/content/drive/MyDrive/Data2.csv")

df.drop('Unnamed: 0', inplace=True, axis=1)
x=df.drop(labels=['Fiyat'],  axis=1)


y=df['Fiyat'].values
x_train, x_test, y_train, y_test=train_test_split(x,y, test_size=0.3,random_state=20)


catb_grid = {
    'iterations': [500,1000],
    'learning_rate': [0.01, 0.1],
    'depth': [3,8] }

catb=CatBoostRegressor()

catb_cv=GridSearchCV(catb, catb_grid,cv=10,n_jobs=-1,verbose=2)
catb_model=catb_cv.fit(x_train,y_train)

predication_test=catb_model.predict(x_test)
```

```python
catb_model.best_estimator_
```

```python
catb_model.best_params_
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test,predication_test))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test, predication_test))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, predication_test)))

df_pred=pd.DataFrame({'Acual value':y_test,'predicted value':predication_test,'Diffrence':y_test-predication_test})

df_pred[0:10]
```
