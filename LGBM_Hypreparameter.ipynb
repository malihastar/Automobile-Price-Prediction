```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
from sklearn import metrics
from lightgbm import LGBMRegressor


df=pd.read_csv("/content/drive/MyDrive/Data2.csv")

df.drop('Unnamed: 0', inplace=True, axis=1)
x=df.drop(labels=['Fiyat'],  axis=1)


y=df['Fiyat'].values
x_train, x_test, y_train, y_test=train_test_split(x,y, test_size=0.3,random_state=20)

lgbm_grid = {
    'colsample_bytree': [0.9,1],
    'learning_rate': [0.1,0.2, 0.5 ] ,
    'n_estimators': [  400,800],
    'max_depth': [12,13,14] }

lgbm = LGBMRegressor()
lgbm_cv_model = GridSearchCV(lgbm, lgbm_grid, cv=10, n_jobs = -1, verbose = 2)


lgbm_model = lgbm_cv_model.fit(x_train, y_train)

```

```python
predication_test = lgbm_model.predict(x_test)

# Evaluating the Algorithm

print('Mean Absolute Error:', metrics.mean_absolute_error(y_test,predication_test))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test, predication_test))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, predication_test)))
```

```python
lgbm_tuned = LGBMRegressor(colsample_bytree=0.7,
                           learning_rate= 0.05,
                           max_depth= 10,
                           n_estimators= 1000)
```
