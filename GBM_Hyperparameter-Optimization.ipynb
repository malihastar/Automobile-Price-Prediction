```python

"""
Created on Fri Oct 29 19:24:21 2021

@author: lenovo
"""
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
from sklearn import metrics
from sklearn.ensemble import GradientBoostingRegressor


df=pd.read_csv("/content/drive/MyDrive/Data2.csv")

df.drop('Unnamed: 0', inplace=True, axis=1)
y=df['Fiyat'].values
x=df.drop(labels=['Fiyat'],  axis=1)

x_train, x_test, y_train, y_test=train_test_split(x,y, test_size=0.3,random_state=20)

params={
 "learning_rate"    : [0.05, 0.5] ,
 "max_depth"        : [ 3, 8],
 'n_estimators'     : [ 200, 1000],
 'subsample'        :[1,0.5,0.75],
}

gbm=GradientBoostingRegressor()

gbm_cv=GridSearchCV(gbm, params, cv=10,n_jobs=-1, verbose=2)

gbm_cv.fit(x_train, y_train)

predication_test=gbm_cv.predict(x_test)



# Evaluating the Algorithm

print('Mean Absolute Error:', metrics.mean_absolute_error(y_test,predication_test))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test, predication_test))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, predication_test)))

print(gbm_cv.best_estimator_)
print(gbm_cv.best_params_)
```
